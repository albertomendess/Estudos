# Rastreando um site completo

```python
pages = set()
```

```python
def getLinks(pageUrl):
    html = urlopen(f"http://en.wikipedia.org{pageUrl}")
    bs = BeautifulSoup(html, 'html.parser')
    for link in bs.find_all('a', href= re.compile('^(/wiki/)')):
        if 'href' in link.attrs:
            if link.attrs['href'] not in pages:
                # Encontamos uma nova página
                newPage = link.attrs['href']
                print(newPage)
                pages.add(newPage)
                getLinks(newPage)
```

## Coletando dados de um site completo

```python
pages = set()
```

```python
def getLinks(pagesUrl):
    html = urlopen(f'http://en.wikipedia.org{pagesUrl}')
    bs = BeautifulSoup(html, 'html.parser')
    try:
        print(bs.h1.get_text())
        print(bs.find(id= 'ca-edit').find('span').find('a').attrs['href'])
    except AttributeError:
        print("This page is missing something! Continuing.")

    for link in bs.find_all('a', href= re.compile('^(/wiki/)')):
        if 'href' in link.attrs:
            if link.attrs['href'] not in pages:
                # Encontramos uma nova página
                newPage = link.attrs['href']
                print('-'*20)
                print(newPage)
                pages.add(newPage)
                getLinks(newPage)
```

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup

html = urlopen('http://en.wikipedia.org/wiki/Python_(programming_language)')
bs = BeautifulSoup(html, 'html.parser')

# Teste básico para `ca-edit`
edit_link = bs.find(id='ca-edit')
if edit_link:
    print("Elemento `ca-edit` encontrado!")
    span = edit_link.find('span')
    if span:
        print("Elemento `<span>` encontrado dentro de `ca-edit`.")
        a_tag = span.find('a')
        if a_tag and 'href' in a_tag.attrs:
            print("Link de edição encontrado:", a_tag.attrs['href'])
        else:
            print("Elemento `<a>` ou atributo `href` não encontrado.")
    else:
        print("Elemento `<span>` não encontrado.")
else:
    print("Elemento `ca-edit` não encontrado!")
```
