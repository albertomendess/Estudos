```python
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN
from sklearn.cluster import MeanShift
import pylab as pl
```

```python
iris = load_iris()
```

# Modelo de aprendizado não supervisionado com K-MEANS

```python
# criando um modelo de aprendizadonão supervisionado com K-means
kmeans = KMeans(n_clusters= 3,
                random_state= 111)
kmeans.fit(iris['data'])
```

```python
kmeans.labels_
```

```python
# reduzindo o numero de campos de 4 pra 2 (algoritmo de redução de dimensionalidade)
pca = PCA(n_components= 2).fit(iris['data'])
pca_2d = pca.transform(iris['data'])
pca_2d.shape
```

```python
pl.figure(figsize= (20, 10))        

for i in range(0, pca_2d.shape[0]):
    if iris['target'][i] == 0:
        c1 = pl.scatter(pca_2d[i,0],
                        pca_2d[i,1],
                        c='r',
                        marker='+')
    elif iris['target'][i] == 1:
        c2 = pl.scatter(pca_2d[i,0],
                        pca_2d[i,1],
                        c='g',
                        marker='o')
    elif iris['target'][i] == 2:
        c3 = pl.scatter(pca_2d[i,0],
                        pca_2d[i,1],
                        c='b',
                        marker='*')

pl.legend([c1, c2, c3], ['Setosa', 'Versicolor', 'Virginica'],
          fontsize= 18,
          loc= 'lower right')
pl.title("Conjunto de dados de treinamento Iris com 3 classes e resultados conhecidos",
         fontsize=24)
pl.axis("off")

pl.show()
```

```python
pl.figure(figsize= (20, 10))
pl.scatter(pca_2d[:, 0], pca_2d[:, 1], c= 'black')
pl.title("Conjunto de dados Iris sem rótulos vizualizado por K-means",
         fontsize= 24)
pl.axis('off')

pl.show()
```

```python
pl.figure(figsize= (20, 10))        

for i in range(0, pca_2d.shape[0]):
    if kmeans.labels_[i] == 1:
        c1 = pl.scatter(pca_2d[i,0],
                        pca_2d[i,1],
                        c='r',
                        marker='+')
    elif kmeans.labels_[i] == 0:
        c2 = pl.scatter(pca_2d[i,0],
                        pca_2d[i,1],
                        c='g',
                        marker='o')
    elif kmeans.labels_[i] == 2:
        c3 = pl.scatter(pca_2d[i,0],
                        pca_2d[i,1],
                        c='b',
                        marker='*')

pl.legend([c1, c2, c3], ['Agrupamento 1', 
                         'Agrupamento 0', 
                         'Agrupamento 2'],
          fontsize= 18,
          loc= 'lower right')
pl.title("3 agrupamentos formado pelo algoritmo K-means",
         fontsize=24)
pl.axis("off")

pl.show()
```

```python
# REPETINDO AS EXECUÇÕES COM UM VALOR K DIFERENTE
kmeans2 = KMeans(n_clusters= 2,
                 random_state= 111)
kmeans2.fit(iris['data'])

pl.figure(figsize= (20, 10))
pl.figure('')
for i in range(0, pca_2d.shape[0]):
    if kmeans2.labels_[i] == 1:
        c1 = pl.scatter(pca_2d[i,0], 
                        pca_2d[i,1], 
                        c= 'r',
                        marker= '+')
    elif kmeans2.labels_[i] == 0:
        c2 = pl.scatter(pca_2d[i,0],
                        pca_2d[i,1],
                        c= 'g',
                        marker= 'o')
        
pl.legend([c1, c2], ['Agrupamento 1', 'Agrupamento 2'],
          fontsize= 20,
          loc= 'upper center')
pl.title('K-means em conjunto de dados Iris com 2 agrupamentos',
         fontsize= 30)
pl.axis('off')

pl.show()
```

```python
# K-MEANS COM 4 AGRUPAMENTOS
kmeans4 = KMeans(n_clusters= 4,
                 random_state= 111)
kmeans4.fit(iris['data'])

pl.figure(figsize= (20, 10))
for i in range(0, pca_2d.shape[0]):
    if kmeans4.labels_[i] == 1:
        c1 = pl.scatter(pca_2d[i,0],
                        pca_2d[i,1],
                        c= 'r',
                        marker= '+')
    elif kmeans4.labels_[i] == 0:
        c2 = pl.scatter(pca_2d[i,0],
                        pca_2d[i,1],
                        c= 'g',
                        marker= 'o')
    elif kmeans4.labels_[i] == 2:
        c3 = pl.scatter(pca_2d[i,0],
                        pca_2d[i,1],
                        c= 'b',
                        marker= '*')
    elif kmeans4.labels_[i] == 3:
        c4 = pl.scatter(pca_2d[i,0],
                        pca_2d[i,1],
                        c= 'c',
                        marker= '^')
        
pl.legend([c1, c2, c3, c4], ['Agrupamento 1',
                             'Agrupamento 2',
                             'Agrupamento 3',
                             'Agrupamento 4'],
          fontsize= 18,
          loc= 'upper center')
pl.title('K-means em conjunto de dados Iris com 4 agupamentos',
         fontsize= 30)
pl.axis('off')

pl.show()
```

```python
# Chamar a função predict com uma única observação
kmeans.predict([[5.1, 3.5, 1.4, 0.2]])
```

```python
kmeans.cluster_centers_
```

```python
# chamar o método predict com um conjunto de pontos de dados
kmeans.predict([[5.1, 3.5, 1.4, 0.2],
                [5.9, 3.0, 5.1, 1.8]])
```

```python
kmeans.labels_
```

```python
#  CRIANDO UM MODELO DE APRENDIZADO NÃO SUPERVISIONADO USANDO O 'DBSCAN'

dbscan = DBSCAN()
# eps= 0.5 --> defaut
# eps= 0.9 --> produz um modelo sem ruído
```

# Modelo de aprendizado não supervisionado com DBSCAN

```python
dbscan.fit(iris['data'])
```

```python
dbscan.labels_
```

```python
# visualizando os agrupamentos
pl.figure(figsize= (20, 10))

pca = PCA(n_components= 2).fit(iris['data'])
pca_2d = pca.transform(iris['data'])
for i in range(0, pca_2d.shape[0]):
    if dbscan.labels_[i] == 0:
        c1 = pl.scatter(pca_2d[i,0],
                        pca_2d[i,1],
                        c= 'r',
                        marker= '+')
    elif dbscan.labels_[i] == 1:
        c2 = pl.scatter(pca_2d[i,0],
                        pca_2d[i,1],
                        c= 'g',
                        marker= 'o')
    elif dbscan.labels_[i] == -1:
        c3 = pl.scatter(pca_2d[i,0],
                        pca_2d[i,1],
                        c= 'b',
                        marker= '*')
        
pl.legend([c1, c2, c3], ['Agrupamento 1', 'Agrupamento 2', 'Ruído'],
          fontsize= 15,
          loc= 'upper center')
pl.title("Agrupamentos e ruído usando o modelo DBSCAN",
         fontsize= 25)
pl.axis('off')

pl.show()
```

```python
ms = MeanShift()
#ms?
```

```python
ms.fit(iris['data'])
```

```python
# VISUALIZANDO OS AGRUPAMENTOS
pca = PCA(n_components= 2).fit(iris['data'])
pca_2d = pca.transform(iris['data'])

pl.figure(figsize= (20, 10))
for i in range(0, pca_2d.shape[0]):
    if ms.labels_[i] == 1:
        c1 = pl.scatter(pca_2d[i,0],
                        pca_2d[i,1],
                        c= 'r',
                        marker= '+')
    elif ms.labels_[i] == 0:
        c2 = pl.scatter(pca_2d[i,0],
                        pca_2d[i,1],
                        c= 'g',
                        marker= 'o')
        
pl.legend([c1, c2], ['Agrupamento 1', 'Agrupamento 2'],
          fontsize= 15,
          loc= 'upper center')
pl.title("Modelo de agrupamento usando MEAN SHIFT",
         fontsize= 25)
pl.axis('off')

pl.show()
```
